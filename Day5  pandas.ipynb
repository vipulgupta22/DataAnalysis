## Multi-Indexing

```python
outer=['Task1','Task1','Task2','Task2']
inner=['Ram','Lucky','Ram','Lucky']
multiIndex=pd.MultiIndex.from_tuples(list(zip(outer,inner)))
multiIndex
```

```python
df3=pd.DataFrame(np.random.rand(4,2),index=multiIndex,columns=['level1','level2'])
```
