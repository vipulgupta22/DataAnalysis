## Concatination

```python
import numpy as np
import pandas as pd
```

```python
df1=pd.DataFrame({'col1':[1,2,7],'col2':[4,5,8],'col3':[3,6,9]})
df1
```

```python
df2=pd.DataFrame({'col1':[11,12,17],'col2':[14,15,18],'col3':[13,16,19]})
df2
```

```python
df3=pd.DataFrame({'col1':[21,22,27],'col2':[24,25,28],'col3':[23,26,29]})
df3
```

```python
pd.concat([df1,df2,df3])
```

```python
a=pd.concat([df1,df2,df3],keys=[1,2,3])
a
```

```python
a.index
```

```python
pd.concat([df1,df2,df3],keys=[1,2])
```

```python
pd.concat([df1,df2,df3]).reset_index()
```

```python
pd.concat([df1,df2,df3]).reset_index(drop=True)
```

```python
pd.concat([df1,df2,df3],ignore_index=True)
```

```python
#prevent the result from including duplicate index
df10=pd.DataFrame({'col1':[100,200,300],'col2':[400,500,600],'col3':[700,800,900]},index=[0,4,5])
```

## Merging

```python
df4=pd.DataFrame({'col1':[1,2,7,10],'col2':[12,4,5,8],'col3':[3,6,9,11]})
df4
```

```python
df5=pd.DataFrame({'col1':[21,2,2,10],'col2':[12,4,35,38],'col3':[33,36,39,311]})
df5
```

```python
pd.merge(df5,df4,on='col2',how='inner')
```

```python
df5.merge(df4,on='col2')
```

```python
pd.merge(df5,df4,on='col2',suffixes=['_df5','_df4'])
```

```python
pd.merge(df5,df4,on='col2',how='outer')
```

```python
pd.merge(df5,df4,on='col2',how='left')
```

```python
pd.merge(df5,df4,on='col2',how='right')
```

```python
df6=pd.DataFrame({'co1':[21,22,2,10],'co2':[2,4,35,38],'co3':[33,6,39,311]})
df6
```

```python
pd.merge(df5,df6,left_on='col2',right_on='co2')
```
